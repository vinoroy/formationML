```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns

sns.set(style="white", color_codes=True)
```

```python
irisData = pd.read_csv('iris.csv')
irisData.head()
```

```python
display(irisData.describe())
```

```python
sns.pairplot(irisData, hue="class", size=3, diag_kind="kde")
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='class',y='petal_length(cm)',data=irisData)
plt.subplot(2,2,2)
sns.violinplot(x='class',y='petal_width(cm)',data=irisData)
plt.subplot(2,2,3)
sns.violinplot(x='class',y='sepal_length(cm)',data=irisData)
plt.subplot(2,2,4)
sns.violinplot(x='class',y='sepal_width(cm)',data=irisData)
```
